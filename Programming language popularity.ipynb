```python
import pandas as pd

# Replace 'filename.csv' with the path to your CSV file
df = pd.read_csv('resources/language_popularity_data.csv')

# Display the first 20 rows
df.head(20)
```

```python
# Set Index
df = df.set_index('Date')

# Identify the top 10 programming languages
top_languages = (df
                .sum()
                .sort_values(ascending=False)
                .head(10))

print(top_languages)

# Filter Dataset by top languages
top_languages_list = top_languages.index.tolist()
top_languages_df = df[top_languages_list]
```

```python
# Plot the top language data
import matplotlib.pyplot as plt

top_languages_df.plot(kind='line', figsize=(12, 8))
plt.title('Popularity of Top 10 Programming Languages Over Time')
plt.xlabel('Date')
plt.ylabel('Popularity')
plt.show()
```

```python
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

# Define plot function
def line_plot_languages(languages):
    # Subset the DataFrame to the selected languages
    df_selected = df[list(languages)]
    
    # Create the plot
    plt.figure(figsize=(15,8))
    for column in df_selected.columns:
        plt.plot(df.index, df_selected[column], label=column)

    plt.xlabel('Date')
    plt.ylabel('Popularity')
    plt.title('Popularity of Programming Languages Over Time')

    # Set xticks
    ticks = plt.xticks()[0]
    labels = df.index[::12]  # Select every 12th label
    plt.xticks(ticks[::12], labels, rotation=45)

    plt.legend()
    plt.show()

    
def pie_plot_languages(languages):
    # Subset the DataFrame to the selected languages
    df_selected = df[list(languages)]
    
    # We'll use the latest data for the pie chart
    recent_data = df_selected.iloc[-1]
    
    # Create the plot
    plt.figure(figsize=(10,10))
    plt.pie(recent_data, labels=recent_data.index, autopct='%1.1f%%', startangle=140)
    plt.title('Popularity of Programming Languages - Most Recent Data')
    plt.show()
```

```python
# Create a multi-select widget for selecting languages
all_languages = df.columns

lang_picker = widgets.SelectMultiple(
    options=all_languages,
    value=[all_languages[0]],
    description='Languages',
    disabled=False
)

# Create a dropdown widget for selecting the plot type
plot_type_picker = widgets.Dropdown(
    options=['Line plot', 'Pie chart'],
    value='Line plot',
    description='Plot type:',
)

# Create a function to handle the interaction between the widgets and the plots
def plot_languages(languages, plot_type):
    if plot_type == 'Line plot':
        line_plot_languages(languages)
    elif plot_type == 'Pie chart':
        pie_plot_languages(languages)
```

```python
# Use the interact function to automatically update the plot when the selections change
interact(plot_languages, languages=lang_picker, plot_type=plot_type_picker);
```
